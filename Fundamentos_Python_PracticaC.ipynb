Mini-práctica C (tipos de datos): pide 3 números al usuario, guárdalos en lista y calcula promedio; muestra el tipo de cada valor antes y después de convertir.

```python

################################################################################
## Diseño y   Construcción:	Elihu Izanami Montoya Estrada
##					izanami49@gmail.com
##					EIME
##	        11 de Febrero del 2026
##          Agenda de Contactos en Python
################################################################################


#Función promedio
def promedio():

    #Lista donde se guaradan los numeros ingresados por el usuario
    valores = []

    #Ciclo inifinito
    while True:
        try:
            #Se pide al uduario ingresar un número entero
            valor = int(input("Ingrese el número: "))
            #Si ingresa un valor igual a 0  sale del ciclo y realiza el promedio
            if valor == 0:
                break
            #Funcion que agrea cada valor a  la lista de valores
            valores.append(valor)
        except ValueError:
            print("Inserta un número.")
    print("Promedio: ", sum(valores)/len(valores))


#Mandando a llamar a la función promedio
promedio()
```
